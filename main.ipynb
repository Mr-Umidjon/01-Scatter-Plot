```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(2, 3)
plt.show()
```

```python
x = [2, 4]
y = [1, 2]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x = [2, 4, '0']
y = [1, 2, 0]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
size = 100
plt.scatter(x, y, s=size)
plt.show()
```

```python
size = [100, 300, 600]
plt.scatter(x, y, s=size)
plt.show()
```

```python
size = [100, 300, 600]
color = ['.1', '.4', '.9']
plt.scatter(x, y, s=size, c=color)
plt.show()
```

```python
size = [100, 300, 600]
color = ['red', '.4', '.9']
plt.scatter(x, y, s=size, c=color)
plt.show()
```

```python
size = [100, 300, 600]
color = ['red', 'b', 'c']
plt.scatter(x, y, s=size, c=color)
plt.show()
```

```python
size = [100, 300, 600]
color = ['red', 'b', '#2f3ccc']
plt.scatter(x, y, s=size, c=color)
plt.show()
```
